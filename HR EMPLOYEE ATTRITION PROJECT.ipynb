```python
from sklearn.preprocessing import LabelEncoder
for column in data.columns:
    if data[column].dtype == np.number:
        continue
    data[column] = LabelEncoder().fit_transform(data[column])
```

```python
data
```

```python
data['Age_Years'] = data['Age']
data = data.drop('Age', axis=1) 
```

```python
data
```

```python
x = data.iloc[:, 1:data.shape[1]].values # .values gives us values in array
y = data.iloc[:, 0].values
print(data.shape)
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=10, criterion = 'entropy', random_state = 0)
forest.fit(x_train, y_train)
```

```python
print(x_train[0:5])
```

```python
score = forest.score(x_train,y_train)
print('randomforest classifier', np.abs(score)*100)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, forest.predict(x_test))

TN = cm[0][0]
TP = cm[1][1]
FN = cm[1][0]
FP = cm[0][1]

print(cm)
print('Model Testing Accuracy is - ',(TP+TN)/(TP+TN+FP+FN))
```

```python
print('Misclassification Rate - ',(FP+FN)/(TP+TN+FP+FN))
print('Precission Rate - ',(TP)/(TP+FP))
print('Recall Rate - ',(TP)/(TP+FN))
```
